### Two options for subplots

```python
#sub-plot option 1
fig, ((ax1, ax2),(ax3,ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10,5))

ax1.plot(x,x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax3.set(ylabel="Price ($)")
ax4.hist(np.random.randn(1000))
```

```python
# Sub-plot option 2
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,5))
ax[0,0].plot(x,x/2)
ax[0,1].scatter(np.random.random(10), np.random.random(10))
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1,0].set(ylabel="Price ($)")
ax[1,1].hist(np.random.randn(1000))
```

```python
unique_make = car_sales["Make"].unique()

def average_stats_per_make(make, car_sales):
    temp_df = car_sales[car_sales["Make"]== make]
    value = temp_df["Odometer (KM)"].sum()/len(temp_df)
    return value

for maker in unique_make:
    car_sales.loc[car_sales['Make'] == maker, 'Average (KM)'] = average_stats_per_make(maker, car_sales)
```

```python
unique_make = car_sales["Make"].unique()
unique_avg = car_sales["Average (KM)"].unique()
unique_make, unique_avg
```

```python
fig,ax = plt.subplots(figsize=(5,5))
ax.bar(unique_make, unique_avg)
ax.set(title="Average Odometer (KMs) per maker", xlabel="Make", ylabel="KMs")
```

## pyplot Vs matplotlib object oriented (OO) method

* When plotting something quickly, use pyplot method
* When plotting advance, use OO method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"]>50]
over_50.head()
```

```python
#Pyplot method
over_50.plot(kind="scatter", x='age', y='chol', c='target')
```

```python
#OO method with pyplot
fig,ax = plt.subplots(figsize=(6,10))
over_50.plot(kind='scatter', x='age', y='chol', c='target', ax=ax)
#ax.set_xlim([45, 100])
```

```python
# OO method from scratch
fig,ax = plt.subplots(figsize=(10,6))

#Plot the data 
scatter = ax.scatter(x = over_50['age'], 
                     y = over_50['chol'], 
                     c = over_50['target'])

#Customize
ax.set(title="Heart Disease and Cholestrol Levels", 
       xlabel='Age',
       ylabel='Cholestrol')

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target")

#Add a horizontal line
ax.axhline(over_50["chol"].mean(), linestyle="--")
```

```python
over_50.head()
```

```python
#Subplot of chol, age, and thalach
fig,(ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10,10))

#Add data to ax0
#Plot the data 
scatter = ax0.scatter(x = over_50['age'], 
                     y = over_50['chol'], 
                     c = over_50['target'])

#Customize
ax0.set(title="Heart Disease and Cholestrol Levels", 
       xlabel='Age',
       ylabel='Cholestrol')

# Add a legend
ax0.legend(*scatter.legend_elements(), title="Target")

#Add a horizontal line
ax0.axhline(y=over_50["chol"].mean(), linestyle="--")

#Add data to ax1
#Plot the data 
scatter = ax1.scatter(x = over_50['age'], 
                     y = over_50['thalach'], 
                     c = over_50['target'])

#Customize
ax1.set(title="Heart Disease and Max Heart Rate", 
       xlabel='Max Heart Rate',
       ylabel='Cholestrol')

# Add a legend
ax1.legend(*scatter.legend_elements(), title="Target")

#Add a horizontal line
ax1.axhline(y=over_50["thalach"].mean(), linestyle="--")


#Add title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold")
```

```python
plt.style.use('seaborn-whitegrid')

# OO method from scratch
fig,ax = plt.subplots(figsize=(10,6))

#Plot the data 
scatter = ax.scatter(x = over_50['age'], 
                     y = over_50['chol'], 
                     c = over_50['target'],
                     cmap = "winter")

#Customize
ax.set(title="Heart Disease and Cholestrol Levels", 
       xlabel='Age',
       ylabel='Cholestrol')

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target")

#Add a horizontal line
ax.axhline(over_50["chol"].mean(), linestyle="--")
```

```python
#Subplot of chol, age, and thalach
fig,(ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10,10))

#Add data to ax0
#Plot the data 
scatter = ax0.scatter(x = over_50['age'], 
                     y = over_50['chol'], 
                     c = over_50['target'],
                     cmap = "winter")

#Customize
ax0.set(title="Heart Disease and Cholestrol Levels", 
       xlabel='Age',
       ylabel='Cholestrol')

ax0.set_xlim([50,80])

# Add a legend
ax0.legend(*scatter.legend_elements(), title="Target")

#Add a horizontal line
ax0.axhline(y=over_50["chol"].mean(), linestyle="--")

#Add data to ax1
#Plot the data 
scatter = ax1.scatter(x = over_50['age'], 
                     y = over_50['thalach'], 
                     c = over_50['target'],
                     cmap = "winter")

#Customize
ax1.set(title="Heart Disease and Max Heart Rate", 
       xlabel='Max Heart Rate',
       ylabel='Cholestrol')

ax1.set_xlim([50,80])
ax1.set_ylim([60,200])

# Add a legend
ax1.legend(*scatter.legend_elements(), title="Target")

#Add a horizontal line
ax1.axhline(y=over_50["thalach"].mean(), linestyle="--")


#Add title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold")
```
